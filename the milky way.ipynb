Problem 2=====================================================

2a) Derive Oort constants (A,B) in terms of rotation curve, $v_c(R)$.

```python
# from http://en.wikipedia.org/wiki/Oort_constnats
from IPython.display import Image
i = Image(filename='Oort_constants_derivation.jpg'); i 
```

Assume the stars are doing circular motions on a plane around the galactic center. 
$$ V_{obs,radial}=V_{star,radial}-V_{sun,radial}=Vcos(\alpha)-V_0sin(l)$$
$$ V_{obs,proper}=V_{star,proper}-V_{sun,proper}=Vsin(\alpha)-V_0cos(l)$$
Circular motion has $v=\Omega r$
$$ V_{obs,radial}=\Omega R cos(\alpha)-\Omega_0 R_0 sin(l)$$
$$ V_{obs,proper}=\Omega R sin(\alpha)-\Omega_0 R_0 cos(l)$$
From the geometry in the above figure
$$ R cos(\alpha)=R_0 sin(l)$$
$$ R sin(\alpha)=R_0 cos(l)-d $$
Take Taylor Expansion
$$ (\Omega-\Omega_0)=(R-R_0)\frac{d\Omega}{dr}|_{R_0}$$
Assume $R-R_0$ is small
$$ R-R_0 = -d cos(l)$$
Writing the velocities in terms of $A$ and $B$
$$ V_{obs,radial}=Adsin(2l)$$
$$ V_{obs,proper}=Adcos(2l)+Bd$$
where
$$ A=-\frac{1}{2} R_0 \frac{d\Omega}{dr}|_{R_0}=\
\frac{1}{2}(\frac{v_c}{R}-\frac{dv_c}{dR})_{R_0}$$
$$ B=-\frac{1}{2} R_0 \frac{d\Omega}{dr}|_{R_0}-\Omega=\
-\frac{1}{2}(\frac{v_c}{R}+\frac{dv_c}{dR})_{R_0}$$

These assumptions are more likely to be valid for stars. Oort constants $A$ measures the shear in the disk at the Sun. $B$ measures the vorticity of the material of the disk.

2b) Derive relation between $\mu_l$ and $l$ in terms of Oort constant.
$$ V_{obs,radial}=Adsin(2l)$$
$$ V_{obs,proper}=Adcos(2l)+Bd$$
Therefore
$$ \frac{V_{obs,proper}}{d}=\mu_l = Acos(2l)+B,\quad  where\ (A,B)\ are\ in\ mas/yr/kpc$$

```python
# copied from http://nbviewer.ipython.org/github/adrn/teaching/blob/master/galactic-dynamics/notebooks/Fit%20a%20line%20to%20data.ipynb
def model(pars, x):
    m,b = pars
    return m*x + b
def error_function(pars, x, y, sigma_y):
    return (y - model(pars, x)) / sigma_y
```

```python
# velocity unit conversion
# Oort constant (A,B) in unit of km/s/kpc, ul in unit of mas/yr
convert = 1.5*10.**8./31536000. # 1au/yr = ? km/s
```

```python
# select stars with circular motion in the galactic plane
index = (abs(b)<10.) & (abs(ub)<(10./convert)) & (abs(ul)<(25./convert)) &\
(abs(distance)<2.) & (abs(el/ul)<0.5)
l, b, pi, ul, ub, epi, el, eb, v, color, v_abs, distance = \
l[index], b[index], pi[index], ul[index], ub[index], epi[index], el[index], eb[index], v[index], color[index], v_abs[index], distance[index]
# oort data
x = np.cos(2.*l/180.*np.pi); y = ul*convert; sigma_y = el*convert
# fit oort
initial_parameter_guess = [1., 0.]
output = optimize.leastsq(error_function, x0=initial_parameter_guess, args=(x, y, sigma_y), full_output=True)
best_pars = output[0]; cov = output[1]
dist = multivariate_normal(mean=best_pars, cov=cov)
# uncertainties
sigma_par=( np.linalg.inv(np.dot(cov.transpose(),cov))/(np.size(x)-2)*\
sum((y-model(best_pars,x))**2) )**0.5
```

```python
# best fit
print(r'$A(km/s/kpc) = %f +/- %f$, $B(km/s/kpc) = %f +/- %f$' \
      %(best_pars[0], sigma_par[0,0], best_pars[1], sigma_par[1,1]))
# plot
plt.figure(figsize=(8,6))
plt.errorbar(x, y, sigma_y, marker='o', linestyle='none', ecolor='b', elinewidth=2, markersize=6)
x_l = np.linspace(min(x), max(x), 100)
param_samples = dist.rvs(size=100)
for params in param_samples:
    plt.plot(x_l, model(params, x_l), marker=None, 
             linestyle='-', color='cyan', alpha=0.2)
plt.plot(x_l, model(best_pars, x_l), marker=None, label='best fit', linestyle='-', color='r', linewidth=1.)
plt.xlabel(r'$cos(2l)$'); plt.ylabel(r'$\mu_l$')
plt.legend(loc='upper left')
```
