
Problem 1: Write a Leapfrog Integrator -----------------------------------------

```python
# leapfrog
def leapfrog_integrate(acceleration_func, x0, v0, dt, nsteps, args=()):
    func = lambda x: acceleration_func(x, *args)
    q, p = np.zeros((np.size(x0),nsteps+1)), np.zeros((np.size(x0),nsteps+1))
    q_0 = x0
    p_0 = v0 - 0.5*dt*func(q_0) # v_{1/2}
    q[:,0], p[:,0] = q_0, p_0
    for i in range(nsteps):
        q_prime = q_0 + dt*p_0
        p_prime = p_0 - 1.0*dt*func(q_prime) # kick-drift-kick
        q[:,i+1], p[:,i+1] = q_prime, p_prime # record
        q_0, p_0 = q_prime, p_prime
        
    return q, p
```


Problem 2: Choose an analytic form for the potential of the disk of the Milky Way ---------------------------------------
$$ \phi (x,y)= \frac{1}{2} {v_0}^2 ln({R_c}^2+x^2+\frac{y^2}{q}) $$
For $R >> Rc$, in 3 dimension, and assume x and y axis are symmetric, this should be adapted to be:
$$ \phi (x,y,z)= \frac{1}{2} {v_0}^2 ln (x^2+y^2+\frac{z^2}{q}) $$ 
The acceleration will be
$$ a_x = {v_0}^2 \frac{x}{x^2+y^2+z^2/q} $$
$$ a_y = {v_0}^2 \frac{y}{x^2+y^2+z^2/q} $$
$$ a_z = {v_0}^2 \frac{z/q}{x^2+y^2+z^2/q} $$
To estimate $v_0$, take the circular speed $v_c$ has
$$ \frac{{v_c}^2}{R} = a_c = \frac{\partial \Phi}{\partial R} = {v_0}^2 \frac{R}{R^2+z^2} $$
where $R^2 = (x^2+y^2)$, since $R>>z$, approximately
$$ v_0 = v_c $$
This potential is simple and consistent with the flat circular-speed curves of many disk galaxies for the outer disk part. This potential will not fit for the inner disk. But since our star sample is in the solar neighborhood, I guess its fine. Assume that we are looking at the outer disk, i.e. with big enough $R$, the rotational velocity at any radius will be constant, i.e. $v_c = v_0$


Observational constraints at the Sun's position:
$$ Distance\ to\ the\ center = 8.5\ kpc = 2.6e17\ km$$
$$ Rotation\ velocity = 220\ km/s $$
The sun is in nearly circular orbit. So the parameter $v_0$ is
$$ v_0 = v_c = 220\ km/s$$ 

```python
# plot rotation curve
radii = np.arange(2,15,1); rot_vel = np.ones(np.size(radii))*220.
plt.plot(radii, rot_vel, 'x')
plt.title('Rotation Curve of the Log Potential')
plt.xlabel(r'$R (kpc)$'); plt.ylabel(r'$V_c (km/s) $')
```


To estimate another parameter in the potential, $q$(axial ratio), I use the fact that the Sun oscillates up and down relative to the Galactic plane approximately 2.7 times per orbit. And the axial ratio that turns out to satisfy this constrain is $q = 0.14$

```python
# log potential
def logpotential_acc(x, weights=np.ones(3)): # num=number of objects
    dim = 3; num = np.size(x)/dim
    x_rearrange = x.reshape(num, dim)

    r_square = np.sum(x_rearrange**2./weights, axis=1) # (R^2+z^2/q)
    r_rearrange = np.repeat(r_square, dim)
    
    acc = 220.**2. * x/np.tile(weights,num) / r_rearrange # (x,y,z/q)/(R^2+z^2/q)
    #acc = 220.**2. * x / np.linalg.norm(x, axis=0)**2. # vknot=220.
    return acc

# main
global kpc, myr
kpc = 3.1*10**16. # km
myr = 3.16e13 # s
d_sun = -8.5*kpc; h_sun = 0.05*kpc; v_sun = 220. # in km, km/s
axis_ratio = 0.14; weights = np.array([1.,1.,axis_ratio])
step_mw = 236
q, p = leapfrog_integrate(logpotential_acc,
                          x0=np.array([d_sun,0,h_sun]), 
                          v0=np.array([0,v_sun,0]),
                          dt=myr, nsteps=step_mw, 
                          args=(weights,))
```

```python
# plot
plt.figure(figsize=(6,6))
plt.plot(0,0,'*', label='MW center')
plt.plot(q[0,0]/kpc,q[1,0]/kpc, 'ro', markersize=10, label='Now')
plt.plot(q[0,:]/kpc,q[1,:]/kpc, 'x', label='Orbit')
plt.xlabel('x [kpc]'); plt.ylabel('y [kpc]')
plt.axis('equal'); plt.legend(loc='upper left')
plt.title("The Sun's orbit in x-y plane in %3.f Myrs" % step_mw)

# plot
plt.figure(figsize=(6,4))
plt.plot(0,0,'*', label='MW center')
plt.plot(q[2,:]/kpc*1000, 'x')
plt.xlabel('T [Myr]'); plt.ylabel('z [pc]')
plt.title("The Sun's orbit in z-direction in %3.f Myrs" % step_mw)
```

Properties of the orbits: These orbits are bound orbits with different kinetic energies oscillating between their lower and upper bounds of R.

```python
x_all = q_all[0:np.size(q_all):3,:]
y_all = q_all[1:np.size(q_all):3,:]
z_all = q_all[2:np.size(q_all):3,:]
R_all = (x_all**2.+y_all**2.)**0.5
```

```python
# calculate orbital period
fft_z = np.fft.fft(z_all)
freq_z = np.fft.fftfreq(nsteps+1, 1)
fft_z_shifted = np.abs(np.fft.fftshift(fft_z))
freq_z_shifted = np.abs(np.fft.fftshift(freq_z))
z_period = 1./freq_z_shifted[np.argmax(fft_z_shifted,axis=1)]

R_mean = np.repeat(np.mean(R_all,axis=1),nsteps+1).reshape(np.size(x),nsteps+1)
R_delmean = R_all-R_mean
fft_R = np.fft.fft(R_delmean)
fft_R_shifted = np.abs(np.fft.fftshift(fft_R))
R_period = 1./freq_z_shifted[np.argmax(fft_R_shifted,axis=1)]

theta_abs = np.arccos(x_all/R_all)/np.pi*180.
theta = theta_abs*(y_all>=0) + (360-theta_abs)*(y_all<0)
t_mean = np.repeat(np.mean(theta,axis=1),nsteps+1).reshape(np.size(x),nsteps+1)
t_delmean = theta-t_mean
fft_t = np.fft.fft(t_delmean)
fft_t_shifted = np.abs(np.fft.fftshift(fft_t))
t_period = 1./freq_z_shifted[np.argmax(fft_t_shifted,axis=1)]
```

```python
plt.figure(figsize=(4,3)) # radial period
plt.hist(R_period)
plt.gca().set_yscale("log"); 
plt.title('orbital periods in the radial direction')

plt.figure(figsize=(4,3)) # azimuthal period
plt.hist(t_period)
plt.gca().set_yscale("log")
plt.title('orbital periods in the azimuthal direction')

plt.figure(figsize=(4,3)) # vertical period
plt.hist(z_period)
plt.gca().set_yscale("log")
plt.title('orbital periods in the vertical direction')
```
