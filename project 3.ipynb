```python
# leapfrog
# NOTE: always feed in cartesian coordinates
def leapfrog_integrate(acceleration_func, x0, v0, dt, t0, nsteps, args=(), n_increase=1000):
    etafinal,omega,a,GM,b=args
    etas = np.linspace(0,etafinal,n_increase+1)
    
    func = lambda x,t: acceleration_func(x,t,etas[0],omega,a,GM,b)
    q, p = np.zeros((np.size(x0),nsteps+1)), np.zeros((np.size(x0),nsteps+1))
    t = np.arange(nsteps+1)*dt+t0
    q_0 = x0
    p_0 = v0 - 0.5*dt*func(q_0,t0) # v_{1/2}
    q[:,0], p[:,0] = q_0, p_0
    for i in range(n_increase):
        func = lambda x,t: acceleration_func(x,t,etas[i+1],omega,a,GM,b)
        q_prime = q_0 + dt*p_0
        p_prime = p_0 + dt*func(q_prime,t[i+1]) # kick-drift-kick
        q[:,i+1], p[:,i+1] = q_prime, p_prime # record
        q_0, p_0 = q_prime, p_prime
        
    func = lambda x,t:acceleration_func(x,t,etas[0],omega,a,GM,b)
    #func = lambda x,t:acceleration_func(x,t,etafinal,omega,a,GM,b)
    for i in range(n_increase,nsteps):
        q_prime = q_0 + dt*p_0
        p_prime = p_0 + dt*func(q_prime,t[i+1]) # kick-drift-kick
        q[:,i+1], p[:,i+1] = q_prime, p_prime # record
        q_0, p_0 = q_prime, p_prime
        
    return t, q, p
```

Problem 1: implement background potential and bar potential

```python
def bg_acceleration(r,GM=1,b=1): # GM=1, b=1
    return 0.-GM*r/(b+(b**2.+r**2.)**0.5)**2./(b**2.+r**2.)**0.5
```

```python
def bg_potential(r): # GM=1, b=1
    return 0.-1./(1.+(1.+r**2.)**0.5)
```

```python
def bar_acceleration(q,t,eta=0.5,omega=0.2,a=2.6): 
    dim = 2; num = np.size(q)/dim
    q_rearrange = q.reshape(num, dim)
    
    r,theta=q_rearrange[:,0],q_rearrange[:,1]
    r1 = r/a
    ar = 0.-(eta*(-0.53/a)*(2.*r1-10.*r1**3.+2.1875*6.*r1**5.-0.65625*8*r1**7.)*(r1<1.)+eta*(-4.*0.017/a)*(r1**-5.)*(r1>=1)) \
        *np.cos(2*(theta-omega*t))
    #at = 0.-(eta*(-0.53/a)*(r1-2.5*r1**3.+2.1875*r1**3.-0.65625*r1**5.)*(r1<1)+eta*(0.017/a)*(r1**-5.) ) \
    at = eta*(-0.53)*(r1**2.-2.5*r1**4.+2.1875*r1**6.-0.65625*r1**8.)*(r1<1.)+eta*(0.017)*(r1**-4.)*(r1>=1.) \
        *2*np.sin(2*(theta-omega*t))
    return np.dstack((ar,at)).reshape(1,np.size(q))
```

```python
def sum_acc(q,t,eta=0.5,omega=0.2,a=2.6,GM=1,b=1): # feed in, give out, cartesian
    dim = 3; num = np.size(q)/dim
    q_rearrange = q.reshape(num,dim)
    x = q_rearrange[:,0]; y = q_rearrange[:,1]; z=q_rearrange[:,2]
    
    r = (x**2+y**2+z**2)**0.5
    xy = (x**2+y**2)**0.5
    theta = ( np.arccos(x/xy)*((y/xy)>=0)+(2*np.pi-np.arccos(x/xy))*((y/xy)<0) )
    phi = ( np.arccos(xy/r)*((z/r)>=0)+(2*np.pi-np.arccos(xy/r))*((z/r)<0) )
    qrt = np.dstack((r,theta)).reshape(1,np.size(r)*2)
    
    ar = bg_acceleration(r,GM,b)+bar_acceleration(qrt,t,eta,omega,a).reshape(num,2)[:,0]
    at = bar_acceleration(qrt,t,eta,omega,a).reshape(num,2)[:,1]
    ap = np.zeros_like(ar)
    
    ax = ar*np.cos(phi)*np.cos(theta)-at*np.sin(theta)
    ay = ar*np.cos(phi)*np.sin(theta)+at*np.cos(theta)
    az = ar*np.sin(phi)
    
    return np.dstack((ax,ay,az)).reshape(1,np.size(q))
```

```python
def bar_potential(q,t,eta=0.5,omega=0.2): # eta=0.5, omega=0.2, a=2.6
    dim = 2; num = np.size(q)/dim
    q_rearrange = q.reshape(num,dim)
    
    r = q_rearrange[:,0]; theta = q_rearrange[:,1]
    
    r1 = r/2.6
    phi = eta*(-0.53)*(r1**2.-2.5*r1**4.+2.1875*r1**6.-0.65625*r1**8.)*(r1<1.)+eta*(0.017)*(r1**-4.)*(r1>=1.)
    return phi*np.cos(2.*(theta-omega*t))
```

```python
def perturb_acc(q,qper,GMper):
    dim = 3; num = np.size(q)/dim
    q_rearrange = q.reshape(num,dim)
    x = q_rearrange[:,0]; y = q_rearrange[:,1]; z=q_rearrange[:,2]
    
    xp,yp,zp = qper[0],qper[1],qper[2]
    
    ax = 0.- GMper / ((x-xp)**2.+(y-yp)**2.+(z-zp)**2.)**1.5 * (x-xp)
    ay = 0.- GMper / ((x-xp)**2.+(y-yp)**2.+(z-zp)**2.)**1.5 * (y-yp)
    az = 0.- GMper / ((x-xp)**2.+(y-yp)**2.+(z-zp)**2.)**1.5 * (z-zp)
    
    return np.dstack((ax,ay,az)).reshape(1,np.size(q))
```

```python
def bg_cartesian(q,GM,b):
    dim = 3; num = np.size(q)/dim
    q_rearrange = q.reshape(num,dim)
    x = q_rearrange[:,0]; y = q_rearrange[:,1]; z=q_rearrange[:,2]
    
    r = (x**2+y**2+z**2)**0.5
    xy = (x**2+y**2)**0.5
    theta = ( np.arccos(x/xy)*((y/xy)>=0)+(2*np.pi-np.arccos(x/xy))*((y/xy)<0) )
    phi = ( np.arccos(xy/r)*((z/r)>=0)+(2*np.pi-np.arccos(xy/r))*((z/r)<0) )
    
    ar = bg_acceleration(r,GM,b)
    
    ax = ar*np.cos(phi)*np.cos(theta)
    ay = ar*np.cos(phi)*np.sin(theta)
    az = ar*np.sin(phi)
    
    return np.dstack((ax,ay,az)).reshape(1,np.size(q))
```

```python
def bg_per_acc(q,qper,GMper,GM,b):
    aper = perturb_acc(q,qper,GMper)
    abg = bg_cartesian(q,GM,b)
    
    a = aper+abg
    
    dim = 3; num = np.size(q)/dim
    a_rearrange = a.reshape(num,dim)
    ax = a_rearrange[:,0]; ay = a_rearrange[:,1]; az=a_rearrange[:,2]
    
    return np.dstack((ax,ay,az)).reshape(1,np.size(q))
```

```python
def leapfrog_integrate2(accelerate_star, accelerate_per, x0, v0, xper0,vper0, dt, t0, nsteps, args=()):
    GM,b,GMper=args
    
    acc_star = lambda x,xper: accelerate_star(x,xper,GMper,GM,b)
    q, p = np.zeros((np.size(x0),nsteps+1)), np.zeros((np.size(x0),nsteps+1))
    t = np.arange(nsteps+1)*dt+t0
    q_0 = x0
    p_0 = v0 - 0.5*dt*acc_star(q_0,xper0) # v_{1/2}
    q[:,0], p[:,0] = q_0, p_0
    
    acc_per = lambda q: accelerate_per(q,GM,b)
    qper, pper = np.zeros((3,nsteps+1)), np.zeros((3,nsteps+1))
    qper_0 = xper0
    pper_0 = vper0 -0.5*dt*acc_per(qper_0)
    qper[:,0], pper[:,0] = qper_0, pper_0
    
    for i in range(nsteps):
        # update star
        q_prime = q_0+dt*p_0
        p_prime = p_0+dt*acc_star(q_prime,qper[:,i])
        q[:,i+1], p[:,i+1] = q_prime, p_prime 
        q_0, p_0 = q_prime, p_prime
        # update perturber
        qper_prime = qper_0 + dt*pper_0
        pper_prime = pper_0 + dt*acc_per(qper_prime)
        qper[:,i+1],pper[:,i+1] = qper_prime, pper_prime
        qper_0, pper_0 = qper_prime, pper_prime
        
    return t, q, p, qper, pper
```

```python
def bound_star_E(q,p,potential,args=()):
    dim = 3; num = np.size(q)/dim
    q_rearrange = q.reshape(num,dim)
    x = q_rearrange[:,0]; y = q_rearrange[:,1]; z = q_rearrange[:,2]
    p_rearrange = p.reshape(num,dim)
    vx, vy, vz = p_rearrange[:,0], p_rearrange[:,1], p_rearrange[:,2]
    
    pot = lambda r: potential(r,*args)
    
    r = (x**2.+y**2.+z**2.)**0.5
    pot_E = pot(r)
    kin_E = 0.5 * (vx**2.+vy**2.+vz**2.)
    
    bound = (pot_E+kin_E)<0
    tot_E = np.sum(pot_E[bound])+np.sum(kin_E[bound])
    return tot_E
```

A real satellite galaxy will change the result, because it has an extended shape and it has gas, stars, and dark matter. 

The shape will change the result because the denser region of the satellite will have stronger effect on the part of the disk galaxy it collides with than the less dense region. 

The different mass component will change the result because although a star has tiny chance to encounter another star in the process, the gas component will definitely run into each other and probably create shocks and get new stars formed. But I am not sure if that counts as dynamical process.
